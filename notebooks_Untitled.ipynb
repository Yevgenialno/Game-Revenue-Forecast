```python
f = open("curves_.txt")
txt = f.readline()
txt = txt[1:len(txt) - 2]
train_rmse_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
train_mape_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
train_mae_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
val_rmse_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
val_mae_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
val_mape_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
test_rmse_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
test_mape_list = list(map(float, txt.split(", ")))
txt = f.readline()
txt = txt[1:len(txt) - 2]
test_mae_list = list(map(float, txt.split(", ")))
```

```python
fig, ax = plt.subplots(nrows = 3, figsize=(10, 10))

ax[0].plot(range(len(train_rmse_list)), train_rmse_list, color = "blue", label = "train")
ax[0].plot(range(len(val_rmse_list)), val_rmse_list, color = "green", label = "validation")
ax[0].plot(range(len(test_rmse_list)), test_rmse_list, color = "red", label = "test")
ax[0].set_title("RMSE")

ax[1].plot(range(len(train_mape_list)), train_mape_list, color = "blue", label = "train")
ax[1].plot(range(len(val_mape_list)), val_mape_list, color = "green", label = "validation")
ax[1].plot(range(len(test_mape_list)), test_mape_list, color = "red", label = "test")
ax[1].set_title("MAPE")

ax[2].plot(range(len(train_mae_list)), train_mae_list, color = "blue", label = "train")
ax[2].plot(range(len(val_mae_list)), val_mae_list, color = "green", label = "validation")
ax[2].plot(range(len(test_mae_list)), test_mae_list, color = "red", label = "test")
ax[2].set_title("MAE")

plt.show()
```
