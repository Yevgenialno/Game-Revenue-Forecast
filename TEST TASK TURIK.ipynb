    CORRELATION MATRIX

```python
# створюємо новий датафрейм різниць деяких з чисельних характеристик основного датафрейму
turik_diffs = pd.DataFrame()
```

```python
# Дані переслані з блокноту Дані (прікол)
turik_diffs['total_sessions_day_diff'] = turik_data.total_sessions_day7 - turik_data.total_sessions_day3 
turik_diffs['chapters_opened_day_diff'] = turik_data.chapters_opened_day7 - turik_data.chapters_opened_day3 
turik_diffs['chapters_finished_day_diff'] = turik_data.chapters_finished_day7 - turik_data.chapters_finished_day3 
turik_diffs['diamonds_spent_day_diff'] = turik_data.diamonds_spent_day7 - turik_data.diamonds_spent_day3 
turik_diffs['tickets_spent_day_diff'] = turik_data.tickets_spent_day7 - turik_data.tickets_spent_day3 
turik_diffs['app_sub_ltv_day_diff'] = turik_data.app_sub_ltv_day3 - turik_data.app_sub_ltv_day1 
turik_diffs['app_iap_ltv_day_diff'] = turik_data.app_iap_ltv_day3 - turik_data.app_iap_ltv_day1 
turik_diffs['ad_ltv_day_diff'] = turik_data.ad_ltv_day3 - turik_data.ad_ltv_day1 
turik_diffs['chapters_opened_session_diff'] = turik_data.chapters_opened_session9 - turik_data.chapters_opened_session3 
turik_diffs['chapters_finished_session_diff'] = turik_data.chapters_finished_session9 - turik_data.chapters_finished_session3 
turik_diffs['diamonds_spent_session_diff'] = turik_data.diamonds_spent_session9 - turik_data.diamonds_spent_session3 
turik_diffs['tickets_spent_session_diff'] = turik_data.tickets_spent_session9 - turik_data.tickets_spent_session3
turik_diffs['diamonds_received_diff'] = turik_data.diamonds_received_day7 - turik_data.diamonds_received_day3
```

```python
# спробуємо побудувати корреляціону матрицю до датафрейму різниць
turik_diffs.corr()
```

```python
corr = turik_diffs.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

title = 'Графік кор. матриці до датафрейму turik_diffs'

plt.figure(figsize=(10, 9))
plt.title(title, fontsize=20)
cmap = sns.diverging_palette(160, 15, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, square=True, linewidths=0.1)
```

```python
# У блокноті з моделлю нема трьох характеристик з цього датафрейму, тому побудуємо іншу корр. матрицю 
turik_diffs_edited = pd.DataFrame()

turik_diffs_edited['total_sessions_day_diff'] = turik_data.total_sessions_day7 - turik_data.total_sessions_day3 
turik_diffs_edited['chapters_opened_day_diff'] = turik_data.chapters_opened_day7 - turik_data.chapters_opened_day3 
turik_diffs_edited['chapters_finished_day_diff'] = turik_data.chapters_finished_day7 - turik_data.chapters_finished_day3 
turik_diffs_edited['chapters_closed_day_diff'] = turik_data.chapters_closed_day7 - turik_data.chapters_closed_day3
turik_diffs_edited['diamonds_received_diff'] = turik_data.diamonds_received_day7 - turik_data.diamonds_received_day3
turik_diffs_edited['diamonds_spent_day_diff'] = turik_data.diamonds_spent_day7 - turik_data.diamonds_spent_day3 
turik_diffs_edited['tickets_spent_day_diff'] = turik_data.tickets_spent_day7 - turik_data.tickets_spent_day3 
turik_diffs_edited['app_sub_ltv_day_diff'] = turik_data.app_sub_ltv_day3 - turik_data.app_sub_ltv_day1 
turik_diffs_edited['app_iap_ltv_day_diff'] = turik_data.app_iap_ltv_day3 - turik_data.app_iap_ltv_day1 
turik_diffs_edited['ad_ltv_day_diff'] = turik_data.ad_ltv_day3 - turik_data.ad_ltv_day1 
```

```python
corr = turik_diffs_edited.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

title = 'Графік кор. матриці до датафрейму turik_diffs_edited'

plt.figure(figsize=(10, 9))
plt.title(title, fontsize=20)
cmap = sns.diverging_palette(160, 15, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, square=True, linewidths=0.1)
```

Dropping unused data

```python
# створюємо новій датафрейм - копію основного датафрейму 
turik = turik_data.drop("install_date", axis=1)
```

FILLING MISSING DATA

```python
# хоча у файлі з моделлю ми викидали клітини із загубленими даними,
# оскільки це найбезпечнішим варіантом,
# у цьому блокноті ми спробуємо відновити їх 
mean_day0 = turik_data["total_sessions_day0"].mean()

mean_day1 = turik_data["total_sessions_day1"].mean()

mean_day3 = turik_data["total_sessions_day3"].mean()

mean_day7 = turik_data["total_sessions_day7"].mean()

turik_incomplete_rows["total_sessions_day0"].fillna(mean_day0, inplace=True)
turik_incomplete_rows["total_sessions_day1"].fillna(mean_day1, inplace=True)
turik_incomplete_rows["total_sessions_day3"].fillna(mean_day3, inplace=True)
turik_incomplete_rows["total_sessions_day7"].fillna(mean_day7, inplace=True)
```

ENCODING MEDIA SOURCES

```python
# розшифровуємо деяки нечисельні характеристики датафрейму використовуючи OrdinalEncoder з модулю sklearn
from sklearn.preprocessing import OrdinalEncoder

media_source = turik_data[["media_source"]]
ordinal_encoder = OrdinalEncoder()
source_encoded = ordinal_encoder.fit_transform(media_source)
```

```python
# після розшифровування отримуємо numpy масив чисел 
type(source_encoded)
```

```python
source_encoded[:10]
```

```python
# Хоча розшіфровані значення не завжди відповідають назвам медіа ресурсів, 
# слід зазначити, що така неточність не завадить процесу аналізу даних  
turik_data["media_source"].head(10)
```

```python
# Зауваження: замість створення нової характеристики розшифрованих значень можна було замінити існуючу.
turik["encoded_media_source"] = source_encoded
```

ENCODING PLATFORM

```python
# Робимо аналогічні до вище описаних диї, але для характеристики платформ користувачів

OS = turik_data[["platform"]]
ordinal_encoder = OrdinalEncoder()
OS_encoded = ordinal_encoder.fit_transform(OS)
```

```python
OS_encoded = OS_encoded.astype(int)

turik["encoded_platform"] = OS_encoded
```
